```python
def left_rotate(T:RBTree, x:Node):
    '''
    Left rotation of node x.
    '''
    y = x.right
    b = y.left
    p = x.parent

    y.parent = p
    if p == T.nil:
        T.root = y #Update the root
    if p.left == x: 
        p.left = y
    else:
        p.right = y
    y.left = x
    x.parent = y
    x.right = b
    if b is not T.nil: #T.nil's parent or children cannot be set manually
        b.parent = x
```

```python
def right_rotate(T:RBTree, y:Node):
    '''
    Left rotation of node x.
    '''
    x = y.left
    b = x.right
    p = y.parent

    x.parent = p
    if p == T.nil:
        T.root = x #Update the root
    if p.left == y: 
        p.left = x
    else:
        p.right = x
    x.right = y
    y.parent = x
    y.left = b
    if b is not T.nil: #T.nil's parent or children cannot be set manually
        b.parent = y
```

```python
def RB_INSERTION_MAINTENANCE(T:RBTree, z:Node):
    '''
    Check and correct the black-red tree properties.
    '''
    while z.parent.color == 0:
        if z.parent == z.parent.parent.left:
            y = z.parent.parent.right
            if y.color == 0:
                z.parent.color = 1
                y.color = 1
                z.parent.parent.color = 0
                z = z.parent.parent
            else:
                if z == z.parent.right:
                    left_rotate(T, z)
                    z = z.parent.parent
                z.parent.parent.color = 0
                z.parent.color = 1
                right_rotate(T, z.parent.parent)
        else:
            y = z.parent.parent.left
            if y.color == 0:
                z.parent.color = 1
                y.color = 1
                z.parent.parent.color = 0
                z = z.parent.parent
            else:
                if z == z.parent.left:
                    right_rotate(T, z)
                    z = z.parent.parent
                z.parent.parent.color = 0
                z.parent.color = 1
                left_rotate(T, z.parent.parent)
```

```python
def RB_INSERTION(T:RBTree, z:Node):
    x = T.root
    y = T.nil

    while x is not T.nil:
        y = x
        if z.key <= x.key:
            x = x.left
        else:
            x = x.right
    
    z.parent = y
    if y == T.nil:
        T.root = z
    else:
        if z.key <= y.key:
            y.left = z
        else:
            y.right = z
    z.color = 0 # Assume 0 is red and 1 is black.
    z.left = T.nil
    z.right = T.nil
    RB_INSERTION_MAINTENANCE(T,z)
```

```python
def RB_TRANSPLANT(T:RBTree, u:Node, v:Node):
    '''
    Replace subtree rooted at u with subtree rooted with v.
    '''
    if u.parent == T.nil:
        T.root = v
    elif u.parent.left == u:
        u.parent.left = v
    else:
        u.parent.right = v
    v.parent = u.parent
```

## Deque

```python
class Deque:
    def __init__(self, nums:list, size:int):
        '''
        nums: the values in the deque
        size: the maximum length of the deque
        '''
        self.nums = nums
        self.size = size
        self.head = 0 # Index of the deque's head
        self.tail = len(self.nums) - 1 # Index of the deque's tail

    def empty(self) -> bool:
        '''
        To see if the deque is empty.
        Returns a bool output indicating whether the deque is empty.
        '''
        if self.tail == -1:
            return True
        else:
            return False
    
    def insertion(self, x:int):
        '''
        x: the element to be inserted

        Insert an element at the head of the deque.
        '''
        if self.tail >= self.size - 1:
            raise Exception('Overflow')
        else:
            self.nums.append(x)
            self.nums[:-1] = self.nums[1:]
            self.nums[0] = x
            self.tail += 1

    def enqueue(self, x:int):
        '''
        x: the element to be put

        Push an element into the tail of the queue.
        '''
        # Overflow detection
        if self.tail >= self.size - 1:
            raise Exception('Overflow')
        else:
            self.nums.append(x)
            self.tail += 1

    def unqueue(self):
        '''
        Remove an element from the head of the queue.
        '''
        if self.empty():
            raise Exception('Underflow')
        else:
            self.nums = self.nums[self.head:]
            self.tail -= 1

    def delete(self):
        '''
        Delete the tail element of the deque.
        '''
        if self.empty():
            raise Exception('Underflow')
        else:
            self.nums = self.nums[:-1]
            self.tail -= 1
```
